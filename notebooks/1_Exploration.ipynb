```python
import os
import matplotlib.pyplot as plt
import cv2
import random
```

```python
#defining path to the dataset (images in our case)
DATA_DIR = '/Users/anish/Documents/xAI for Medical Diagnosis/data/raw/'
CLASS = ['COVID', 'Lung_Opacity', 'Viral Pneumonia', 'Normal']
```

```python
#counting number of images in each class
for cls in CLASS:
    print(f"Class: {cls}")

    img_dir = os.path.join(DATA_DIR, cls, "images")
    mask_dir = os.path.join(DATA_DIR, cls, "masks")

    #preview first 5 samples
    image_files = os.listdir(img_dir)[:5]

    #creating figure and set of subplots
    fig, axes = plt.subplots(len(image_files), 2, figsize=(8, 4 * len(image_files))) #(rows, columns, figsize(width, height with number of images))

    for i,fname in enumerate(image_files):

        #current path of image and mask
        img_path = os.path.join(img_dir, fname)
        mask_path = os.path.join(mask_dir, fname)

        #reading image and mask
        img = cv2.imread(img_path)
        mask = cv2.imread(mask_path,cv2.IMREAD_GRAYSCALE) #mask is grayscale/single channel

        if img is not None and mask is not None:
            #converting BGR to RGB
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            #plotting image and mask
            axes[i, 0].imshow(img)
            axes[i, 0].set_title(f"Image: {fname}")
            axes[i, 0].axis('off')

            axes[i, 1].imshow(mask, cmap='gray')
            axes[i, 1].set_title(f"Mask: {fname}")
            axes[i, 1].axis('off')
        else:
            axes[i, 0].text(0.5, 0.5, 'Image or mask not found', horizontalalignment='center', verticalalignment='center')
            axes[i, 1].axis('off')

    plt.tight_layout()
    plt.show()
            



       
```
